```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage.filters import threshold_sauvola 
from skimage.color import rgb2gray
from skimage.io import imread

img = cv2.imread('/Users/muthukrishnan/programming/ml_research/resources/dummy_pans/pan_new/Keaton_Bates.jpg');
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def getMSER_Regions(img):
    mser = cv2.MSER_create()
    return mser.detectRegions(img)

regions, boxes = getMSER_Regions(img)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.imshow(img)
for box in boxes:
    minr = box[0]
    minc = box[1]
    maxr = box[2]
    maxc = box[3]
    rect = mpatches.Rectangle((minr, minc), maxr - minr,maxc - minc,
                              fill=False, edgecolor='red', linewidth=2)
    ax.add_patch(rect)
    
ax.set_axis_off()
ax.plot(boxes[:,0],boxes[:,1])
plt.tight_layout()
plt.show()
```
